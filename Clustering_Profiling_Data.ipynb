**SWAPPING+DELIVERY**

```python
df=data.iloc[:,8:13]
```

```python
df
```

```python
arr=df.values
```

```python
arr
```

```python
kproto=KPrototypes(n_clusters=2,verbose=2,max_iter=20)
```

```python
clusters=kproto.fit_predict(arr,categorical=[0,2,3,4])
```

```python
print(kproto.cluster_centroids_)
```

```python
cluster_dict=[]
for i in clusters:
  cluster_dict.append(i)
```

```python
cluster_dict
```

```python
df['Category']=cluster_dict
```

```python
df
```

```python
df[df['Category']==0]
```

```python
df[df['Category']==1]
```

```python
df[df['Category']==0].to_csv('Swap+Delivery1.csv')
```

```python
df[df['Category']==1].to_csv('Swap+Delivery2.csv')
```
